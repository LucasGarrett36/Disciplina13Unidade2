```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

dados = {
    "Dia": ["Seg", "Ter", "Qua", "Qui", "Sex", "Sáb", "Dom"],
    "Vendas": [1500, 1800, 2200, 2100, 2600, 3100, 2800],
    "Clientes": [50, 65, 70, 80, 90, 120, 110],
    "Lucro": [400, 500, 700, 750, 900, 1200, 1100]
}

df = pd.DataFrame(dados)

sns.set_style("whitegrid")

# --- Gráfico 1: Barras (Total de Vendas por Dia) ---
plt.figure(figsize=(8, 5))
sns.barplot(x="Dia", y="Vendas", data=df, palette="Blues_d")
plt.title("Total de Vendas por Dia", fontsize=14)
plt.xlabel("Dia da Semana", fontsize=12)
plt.ylabel("Total de Vendas (R$)", fontsize=12)
plt.show()

# --- Gráfico 2: Dispersão (Clientes vs. Vendas) ---
plt.figure(figsize=(8, 5))
sns.scatterplot(x="Clientes", y="Vendas", data=df, color="red", s=100)
plt.title("Relação entre Número de Clientes e Vendas", fontsize=14)
plt.xlabel("Número de Clientes", fontsize=12)
plt.ylabel("Total de Vendas (R$)", fontsize=12)
plt.show()

# --- Gráfico 3: Heatmap (Correlação entre Vendas, Clientes e Lucro) ---
plt.figure(figsize=(6, 4))
sns.heatmap(df[["Vendas", "Clientes", "Lucro"]].corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=1)
plt.title("Matriz de Correlação entre Variáveis", fontsize=14)
plt.show()
```
